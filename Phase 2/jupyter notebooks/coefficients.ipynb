# Try and create some data with a correlation coefficient of zero.

Create two pandas series objects, each with random numbers as entries.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) #set a seed for reproducibility, this ensures that the sequence of random numbers generated will be the same everytime the code is executed with the same seed value.
random_1 = pd.Series(np.random.rand(100)) #First series with random values
random_2 = pd.Series(np.random.rand(100)) #Second series with random values
```

Plot a scatter diagram of this random data

```python
plt.scatter(random_1, random_2)
plt.xlabel("Random 1")
plt.ylabel("Random 2")
plt.title("Scatter diagram of random data")
plt.show()
```

Calculate the correlation coefficient of your two series

```python
corr_coeff = random_2.corr(random_1)
print(f"Correlation Coefficient is: {corr_coeff}")
```

EXERCISE:
You will still need two pandas series objects. The first will be full of random numbers again, some positive and some negative. But fill the second with the square of each element in the first. This way we will hard-wire a relationship into the two lists of numbers. We want to get even closer to zero! It sounds counter-intuitive, but...

Find the correlation coefficient for your set of data.

```python
series_1 = pd.Series(np.random.randn(1000)) #randn ensures there are a mixture of positive and negative numbers
series_2 = pd.Series(series_1**2) #This will generate the values of the first series squared

plt.scatter(series_1, series_2)
plt.xlabel("Series 1")
plt.ylabel("Series 2")
plt.title("Scatter diagram of squared random data")
plt.show()

correlation_coeff = series_1.corr(series_2)
print(f"Correlation Coefficient is: {correlation_coeff}")
```
