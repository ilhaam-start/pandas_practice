# Aiming for a correlation coefficient of zero

## Using random numbers

```python
from random import uniform
import pandas as pd

def random_nums(n):
    out = []
    while n > 0:
        n -= 1
        out.append(uniform(10, 20))
    return out

list_a = pd.Series(random_nums(200))
list_b = pd.Series(random_nums(200))

data = pd.DataFrame({
    "random_nums_a": list_a,
    "random_nums_b": list_b,
})

data.plot.scatter(x = "random_nums_a", y = "random_nums_b", title="random numbers vs random numbers")

print("The random numbers have correlation coefficient: ", list_a.corr(list_b))
```

```python
import matplotlib.pyplot as plt
import numpy as np

dice = {
    "scores on red": [1, 4, 3, 2, 3, 6],
    "scores on green": [1, 3, 6, 2, 2, 5]
}

dice_df = pd.DataFrame(dice)

x_val = dice_df['scores on red']
y_val = dice_df['scores on green']

correlation_coeff = x_val.corr(y_val)

a, b = np.polyfit(x_val, y_val, 1)
line_of_best_fit = a * x_val + b

plt.scatter(x_val, y_val)
plt.plot(x_val, line_of_best_fit, color='red')
plt.xlabel("Scores on red")
plt.ylabel("Scores on green")
plt.title("Scatter graph of Scores on red vs Scores on green")
plt.grid(True)
plt.show()
```

```python
print("Correlation Coefficient:", correlation_coeff)
```
