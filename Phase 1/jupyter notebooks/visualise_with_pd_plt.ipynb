```python
march_dict_scorecard = {
    "Will" : [8, 4, 3, 3, 4, 4, 2, 3, 4, 4, 7, 4, 3, 3, 4, 3, 2, 3],
    "Amira" : [3, 2, 2, 3, 4, 7, 5, 3, 5, 2, 5, 2, 4, 5, 4, 3, 4, 3],
    "Steve" : [2, 4, 2, 4, 5, 3, 2, 5, 2, 5, 3, 5, 4, 4, 3, 2, 6, 3]
    }

march_score_frame = pd.DataFrame(march_dict_scorecard, index=range(1, 19))

march_score_frame
```

```python
april_dict_scorecard = {
"Will" : [3, 4, 2, 3, 4, 4, 2, 2, 3, 4, 4, 6, 3, 2, 4, 3, 3, 3],
"Amira" : [3, 4, 2, 3, 4, 3, 5, 3, 5, 2, 4, 2, 4, 5, 4, 3, 4, 3],
"Steve" : [3, 2, 3, 4, 5, 3, 2, 5, 2, 5, 2, 6, 4, 4, 3, 2, 4, 3]
}

hole_comprehension = [f"Hole {i}" for i in range(1, 19)]
hole_index = pd.Index(hole_comprehension, name='Course Hole')
# A list comprehension to add the String "Hole" to each Course Hole Index
april_score_frame = pd.DataFrame(april_dict_scorecard, index=hole_index)

x = hole_index
y1 = march_score_frame['Will']
y2 = april_score_frame['Will']
# Two Y axes, the first is Will's score in March, the second is his score in April

plt.plot(x, y1, label="March")
plt.plot(x, y2, label="April")
# Two plots for each line

plt.xticks(rotation=-45) #rotation of the names on the y-axis
plt.yticks(np.arange(1, 8, step=1)) #y-axis start at 1, end at 8, go up in 1 intervals
# Some tidying so that the hole names don't overlap, and the score is in the right range

plt.xlabel("Course Hole")
plt.ylabel("Shots Taken (Lower Better)")
# Labelling the axes

plt.legend(loc="upper right")
# Adding a legend 

plt.show()
```

```python
average_march = march_score_frame.sum()/18
average_april = april_score_frame.sum()/18

x = ["March", "April"]
y = [average_march, average_april]

# Two axes, x is the month, y is their average scores for March and April

plt.plot(x, y, label=["Will", "Amira", "Steve"])
# one plot for the line

plt.xticks(rotation=-45) #rotation of the names on the y-axis
plt.yticks(np.arange(3, 4, step=0.25)) #y-axis start at 3, end at 4, go up in 0.25 intervals
# Some tidying so that the hole names don't overlap, and the score is in the right range

plt.xlabel("Month")
plt.ylabel("Average Score")
# Labelling the axes

plt.legend(loc="upper right")
# Adding a legend 

plt.show()
```

```python
hap15 = pd.read_csv("2015.csv")
hap16 = pd.read_csv("2016.csv")
hap17 = pd.read_csv("2017.csv")
hap18 = pd.read_csv("2018.csv")

hap18.columns
```

# What Statistical Data Types are Happiness Rank and GDP per capita? Explain in your notebook.

```python
hap18['GDP per capita'].head()
```

GDP per capita is an example of Continuous data. It can take any value between a range, and is not a count, unlike...

```python
hap18['Overall rank'].head()
```

Overall rank or Happiness Rank, is a an example of Discrete data, as it is a count. It cannot be a float - a country cannot be Rank 1.4, or 2.3. They are either first, or second etc.

```python
y1 = [
hap15.loc[hap15['Country']=='United Kingdom']['Happiness Rank'],
hap16.loc[hap16['Country']=='United Kingdom']['Happiness Rank'],
hap17.loc[hap17['Country']=='United Kingdom']['Happiness.Rank'],
hap18.loc[hap18['Country or region']=='United Kingdom']['Overall rank']]

for i in y1:
    print(i)
```

```python
y2 = [
hap15.loc[hap15['Country']=='United Kingdom']['Economy (GDP per Capita)'],
hap16.loc[hap16['Country']=='United Kingdom']['Economy (GDP per Capita)'],
hap17.loc[hap17['Country']=='United Kingdom']['Economy..GDP.per.Capita.'],
hap18.loc[hap18['Country or region']=='United Kingdom']['GDP per capita']]

for i in y2:
    print(i)
```

```python
x = [
    '2015', '2016', '2017', '2018'
]
```

```python
plt.plot(x, y1)
plt.xlabel('Year')
plt.ylabel('Happiness Rank')
```

```python
plt.plot(x, y2)
plt.xlabel('Year')
plt.ylabel('GDP')

plt.show()
```

# Can you plot both the GDP axis and the Happiness axis on the same chart to show the similarities?

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel('Year')
ax1.set_ylabel('Happiness Rank', color='r')
ax2.set_ylabel('GDP', color='b')
```
